```python
plt.figure(figsize=(11,7))
# 输出的是credit.policy值为1的这列数据
loans['credit.policy']==1
# 输出的是图表化的值为1的所有数据
loans[loans['credit.policy']==0]
# 将值为0的数据画一个直方图，bins参数的意思是每个直方图的单个小图的长度，也就是横向的间距
loans[loans['credit.policy']==0]['fico'].hist(alpha=0.5,color='red',bins=20,label='Credit.Policy=0')
```

```python
# Create a histogram of two FICO distributions on top of each other, one for each credit.policy outcome
plt.figure(figsize=(11,7))
# 这个语句是横坐标是借款人的fcio评分与符合信贷标准的人之间的分布情况，可以看出信用度在680-720之间的人数最多，最高达到了820人
loans[loans['credit.policy']==1]['fico'].hist(alpha=0.5,color='blue',
                                              bins=20,label='Credit.Policy=1')
# 这个语句是借款人的fico评分与不符合信贷标准的人之间的分布情况，信用度在650-680之间的人数最多，但也只有200人上下
loans[loans['credit.policy']==0]['fico'].hist(alpha=0.5,color='red',
                                              bins=20,label='Credit.Policy=0')
# 给图形加上图例，也就是右上角那个标注蓝色的是credit=1，红色是credit=0的那个
plt.legend()
plt.xlabel('FICO')
```

```python
# 如果只是单一的loans[],这样出来的就是如下的情况，把credit作为纵坐标，也就是因变量，其他的那些列的数据作为横坐标，也就是变量画图。
loans[loans['credit.policy']==1].hist(alpha=0.5,color='blue',
                                              bins=40,label='Credit.Policy=1')
```

```python
# Create a similar figure, except this time select by the not.fully.paid column
# 现在也能看出，在信用度为680-720之间的有能力偿还和没有能力偿还的人的人数是最多的，大概在七八百之间，不过不连续
plt.figure(figsize=(11,7))
loans[loans['not.fully.paid']==1]['fico'].hist(alpha=0.5,color='yellow',bins=30,label='not.fully.paid = 1')
loans[loans['not.fully.paid']==0]['fico'].hist(alpha=0.5,color='red',bins=30,label='not.fully.paid = 0')
plt.legend()
plt.xlabel('FICO')
```

```python
# Create a countplot using seaborn showing the counts of loans by purpose, with the color hue defined by not.fully.pai
plt.figure(figsize=(11,7))
loans[loans['not.fully.paid']==1]['purpose'].hist(alpha=0.3,bins=5,color='blue',label='not.fully.paid = 1')
loans[loans['not.fully.paid']==0]['purpose'].hist(alpha=0.7,bins=5,color='orange',label='not.fully.paid = 0')
plt.legend()
plt.xlabel('purpose')
```

```python
plt.figure(figsize=(11,7))
# 这个函数调用的就是，x的值是横坐标想要显示的值,也就是这一栏下的几个分类的详细的分布情况
sns.countplot(x='purpose', data=loans)
```

```python
# 基于上面那个图像，再将这些图中，每个column里面，按照not.fully paid的值进行分类,
# palette 是调色板的意思，后面那个set1就是参数，可以换成set3之类的
# 这一语句必须每一次画图的时候都要带上，否则图形画出来就是默认大小，很影响可视化的大小
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=loans,palette='Set2')
# 从这个图形上看来，not.fully.paid=0的人主要的借款目的是debt_consolidation，高达3300人，当然值为1的也是这个目的，只不过数值没有为0的大
```

```python
# 要想放在一个水平上的话，要使用的是 col
#这里的x，y就是图表上的横纵坐标,col的意思就是将这个图按照这个分裂成为两个图，hue的意思是在这两个分类的图中，再按照hue的参数
# 进行区分，shareX,Y的意思要不要共享X,Y轴的值，因为有些值的取值范围是不一样的，这里我是共享的，最后那个palette是给图形设置样式的
sns.lmplot(x='fico',y='int.rate',data=loans,col='not.fully.paid',hue='credit.policy',sharey=True,sharex=True,palette="husl")
```

```python
# 如果这里的分类是用的row的话，就是竖直的展示
sns.lmplot(x='fico',y='int.rate',data=loans,row='not.fully.paid',hue='credit.policy',sharey=True,sharex=True,palette="husl")
```

```python
# 将purpose的值变成单独的行，将原来的purpose哪一行删除， drop_first=True  并且在数据后面添加上purpose的值的列，变成新列。
cat_feats = ['purpose']
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
# 最后可以看出，数据中出现了新加的几列
final_data.info()
```

```python
# Use sklearn to split your data into a training set and a testing set as we've done in the past.
from sklearn.model_selection import train_test_split
# drop函数默认删除行，列需要加axis = 1
# 这里就是讲前面的这些除了notfully paid的列保存为X集合，
X = final_data.drop('not.fully.paid',axis=1)
```

```python
X
```

```python
y = final_data['not.fully.paid']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
dt_predictions = dtree.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,dt_predictions))
print(confusion_matrix(y_test,dt_predictions))
```

```python
target_names = ['not.fully.paid=1', 'not.fully.paid=0']
dt_mat = confusion_matrix(y_test, dt_predictions)
plt.title("predict-to-true confusion matrix")
sns.heatmap(dt_mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=target_names,
         yticklabels=target_names)
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.show()
```

```python
# Training the Random Forest model
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train,y_train)
rfc_predictions = rfc.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,rfc_predictions))
print(confusion_matrix(y_test,rfc_predictions))
```

```python
target_names = ['not.fully.paid=1', 'not.fully.paid=0']
rfc_mat = confusion_matrix(y_test, rfc_predictions)
plt.title("predict-to-true confusion matrix")
sns.heatmap(rfc_mat.T, square=True, annot=True, fmt='d', cbar=True, xticklabels=target_names,
         yticklabels=target_names)
plt.xlabel('rfc true label')
plt.ylabel('rfc predicted label')
plt.show()
```
